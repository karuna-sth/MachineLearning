```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv("datas/iris.csv")
```

```python
iris.head(10)
```

```python
iris.info() #gives all information on data
```

#### Form this we can know that the dataset has 150 rows and 5 columns. And name of columns. which can also be shown using 'iris.shape' 
#### which give dimension of the dataset and 'iris.columns' which gives the name of columns.

```python
print(iris.variety.value_counts())
```

#### Here, value_counts() gives number of different values in an clolumns. So, now we know that we have 50 datas fo each class of iris flower i.e 
#### Virginica, Setosa and Versicolor

```python
iris.plot(kind='scatter', x = 'sepal.length', y='sepal.width')
plt.show()
```

plotting using pandas where features in x axis is sepal.length and y axis is sepal.width for all class but we cannot distingiush between sepal.length and sepal.width of different class 

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='variety', size = 6).map(plt.scatter, "sepal.length", "sepal.width").add_legend()
plt.show()
```

we use seaborn plot to distingiush the species by color hue='variety' it gives different color to different variety in plot and add_legend() is used to add index in graph
#### Observations
1. Using features sepal.length and sepal.width we can distingiush Setosa from other two
2. But cannot distingiush between virginica and versicolor as they have considerable overlap 

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue='variety', height=3)
plt.show()
```

From this Observation we can see that
1. Plot between petal.length and petal width is best way to distinguish 3 species
2. Also, in the same plot setosa can be identified i.e linearly seperably but virginica and versicolor have some overlap i.e almost lineraly seperable
3. So, to classify the species we can from simple model by if-else condition by finding seperable lines.
